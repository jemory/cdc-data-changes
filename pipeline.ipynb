# imports

```python
import requests
import pandas as pd
```

# acquire data

```python
api_url = "https://data.cdc.gov/api/views.json"
response = requests.get(api_url)
data = response.json()
```

```python
data[0:4]
```

```python
df = pd.DataFrame(data)
```

# clean data

```python
backup_df = df.copy()
```

```python
df.dtypes
```

# analyze data

```python
just_names = df[['id', 'name']]
```

```python
just_names
```

```python
just_names = just_names.sort_values('id')
```

```python
just_names
```

# export data

```python
df.to_csv('exported_data/cdc-data.csv')
df.to_json('exported_data/cdc-data.json')
just_names.to_csv('exported_data/cdc-data-names.csv')
just_names.to_json('exported_data/cdc-data-names.json')
```
